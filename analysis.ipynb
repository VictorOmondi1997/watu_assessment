```python
import pandas as pd
```

```python
client = pd.read_csv('./data/client.csv', sep='|')
loan = pd.read_csv('./data/loan.csv', sep='|')
vehicle = pd.read_csv('./data/vehicle.csv', sep='|')
```

```python
client.head()
```

```python
client.columns
```

```python
client.drop(columns=['Unnamed: 0', 'Unnamed: 6'], inplace=True)
client.head()
```

```python
client.drop([0], inplace=True)
client.head()
```

```python
loan.head()
```

```python
loan.drop(columns=['Unnamed: 0', 'Unnamed: 6'], inplace=True)
loan.drop([0], inplace=True)
loan.head()
```

```python
vehicle.head()
```

```python
vehicle.drop(columns=['Unnamed: 0', 'Unnamed: 4'], inplace=True)
vehicle.drop([0], inplace=True)
vehicle.head()
```

```python
def trim(col):
    return col.str.strip()

client=client.apply(trim)
loan = loan.apply(trim)
vehicle = vehicle.apply(trim)
```

```python
client.head()
```

```python
loan.head()
```

```python
vehicle.head()
```

```python
loan.dtypes
```

```python
client['client_id'] = client.client_id.astype(int)
client['date_of_birth'] = pd.to_datetime(client.date_of_birth)
client.head()
```

```python
loan['loan_id'] = loan.loan_id.astype(int)
loan['client_id'] = loan.client_id.astype(int)
loan['vehicle_id'] = loan.vehicle_id.astype(int)
loan['principal_amount'] = loan.principal_amount.astype(int)
loan['submitted_on_date'] = pd.to_datetime(loan.submitted_on_date)
loan.head()
```

```python
vehicle['vehicle_id'] = vehicle.vehicle_id.astype(int)
vehicle.head()
```

```python
client.to_csv('data/clean/client.csv', index=False)
loan.to_csv('data/clean/loan.csv', index=False)
vehicle.to_csv('data/clean/vehicle.csv', index=False)
```
